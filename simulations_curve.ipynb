```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.style.use('seaborn-darkgrid')
df6=pd.read_csv("advd.csv")
df8=pd.read_csv("advd_exp.csv")
pal=['darkgoldenrod','dodgerblue','fuchsia', 'slateblue']
# pal=['darkgoldenrod','dodgerblue']
ax=sns.lineplot(data=df6, x="x", y="y", hue=df6['hue'], palette=pal,marker='.', linestyle=':', linewidth=3)
# ax=sns.lineplot(data=df6, x="x", y="y", hue=df6['hue'], marker='.', linestyle=':')
ax=sns.lineplot(data=df8, x="x", y="y", color="red", linewidth=3)
ax.invert_xaxis()
# plt.savefig('adv.png', dpi=300, bbox_inches='tight')
```
