# Plotting mean speed

```python
import pandas as pd
import numpy as np;
import seaborn as sns; 

import matplotlib.pyplot as plt 


df= pd.read_csv("P1Mu.csv") #mporting data as a dataframe from a .csv file
x=df.mu #setting the X value

plt.style.use('seaborn-darkgrid') #choosing the plot style

g=sns.displot(df,x=x,  color='dimgray', binwidth=3) #plotting a bar graph

g.set(xlim=(10,45)) #set the X-axis limit accordingly

g=plt.plot(x, np.full_like(x, -0.01), '|k', markeredgewidth=1, color='black') #plotting the distribution as a rug plot
sns.displot(df,x=x, color="dimgrey",kind="ecdf"  ) # Plotting a univariate distribution along the x axis
#plt.savefig('Mu_harvfinal.png', dpi=300, bbox_inches='tight')
```
